#### Define data.

```python
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

#### There are 2 ways of doing this. Do in both ways.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

plt.style.use('ggplot')

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(5, 5), sharex=True)
ax[0].plot(x, y, color='b', ls='--')
ax[1].plot(x, z, color='r', ls='-')

plt.show()
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here

plt.style.use('ggplot')
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 5), sharex=True)
ax[0].plot(x, y)
ax[0].set_xlabel('x')
ax[0].set_xlabel('y')

ax[1].plot(x, z)
ax[1].set_xlabel('x')
ax[1].set_xlabel('z')

plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax[0].plot(x, x**2, np.exp(x))
axes[0].set_title("Normal scale")

# modificando la escala para ambos ejes
ax[0].set_yscale('linear')
ax[0].set_xscale('linear')

ax[1].plot(x, x**2, np.exp(x))
ax[1].set_yscale('log')
ax[1].set_xscale('log')
axes[1].set_title("Logarithmic scale (y)");
```
